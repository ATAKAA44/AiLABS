```python
# nmbr of missed values for every colmn
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
# fillnaa
homeplanet_mode = df['HomePlanet'].mode()[0]
cryosleep_mode = df['CryoSleep'].mode()[0]
vip_mode = df['VIP'].mode()[0]
name_mode = df['Name'].mode()[0]

roomservice_median = df['RoomService'].median()
foodcourt_median = df['FoodCourt'].median()
shoppingmall_median = df['ShoppingMall'].median()
spa_median = df['Spa'].median()
vrdeck_median = df['VRDeck'].median()


# suggested by JL compiler
df['HomePlanet'] = df['HomePlanet'].fillna(homeplanet_mode)
df['CryoSleep'] = df['CryoSleep'].fillna(cryosleep_mode).astype(str)
df['VIP'] = df['VIP'].fillna(vip_mode).astype(str)
df['Name'] = df['Name'].fillna(name_mode)

df['RoomService'] = df['RoomService'].fillna(roomservice_median)
df['FoodCourt'] = df['FoodCourt'].fillna(foodcourt_median)
df['ShoppingMall'] = df['ShoppingMall'].fillna(shoppingmall_median)
df['Spa'] = df['Spa'].fillna(spa_median)
df['VRDeck'] = df['VRDeck'].fillna(vrdeck_median)

missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
# normalizacia
scaler1 = MinMaxScaler()
scaler2 = MinMaxScaler()
scaler3 = MinMaxScaler()
scaler4 = MinMaxScaler()
scaler5 = MinMaxScaler()

df['Age'] = scaler1.fit_transform(df[['Age']])
df['RoomService'] = scaler2.fit_transform(df[['RoomService']])
df['FoodCourt'] = scaler3.fit_transform(df[['FoodCourt']])
df['Spa'] = scaler4.fit_transform(df[['Spa']])
df['ShoppingMall'] = scaler5.fit_transform(df[['ShoppingMall']])

df.head(10)
```

```python
# categoricaldata transforming
df = pd.get_dummies(df, columns=['Cabin'], drop_first=True)

df.head(len(df))
```

```python
df.to_csv("train2.csv", index=False)
```
