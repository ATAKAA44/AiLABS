```python
# nmbr of missed values for every colmn
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
# fillnaa
age_median = df['Age'].median()

cabin_mode = df['Cabin'].mode()[0]
embarked_mode = df['Embarked'].mode()[0]


# suggested by JL compiler
df['Age'] = df['Age'].fillna(age_median)

df['Cabin'] = df['Cabin'].fillna(cabin_mode)
df['Embarked'] = df['Embarked'].fillna(embarked_mode)


missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
# normalizacia
scaler1 = MinMaxScaler()
scaler2 = MinMaxScaler()
scaler3 = MinMaxScaler()
scaler4 = MinMaxScaler()


df["Pclass"] = scaler1.fit_transform(df[["Pclass"]])
df["SibSp"] = scaler2.fit_transform(df[["SibSp"]])
df["Parch"] = scaler3.fit_transform(df[["Parch"]])
df["Fare"] = scaler4.fit_transform(df[["Fare"]])


df.head(10)
```

```python
df["Cabin"] = df["Cabin"].str[0]


df.head(len(df))
```

```python
# categoricaldata transforming
df = pd.get_dummies(df, columns=['Sex'], drop_first=True)
df = pd.get_dummies(df, columns=['Cabin'], drop_first=True)
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)


df.head(len(df))
```

```python
df.to_csv("train.lab1.new.csv", index=False)
```
